#### Importar dados

```python
import pandas as pd
```

#### Carga dos dados

```python
dfmat = pd.read_csv(r"C:\Users\Yago\Downloads\GIT\ProjetoEscola\data_student-mat.csv", delimiter= ';', header=0, index_col=False)
dfpor = pd.read_csv(r"C:\Users\Yago\Downloads\GIT\ProjetoEscola\data_student-por.csv", delimiter= ';', header=0, index_col=False)
```

#### Visualização geral dos dados

```python
dfmat.head()
```

#### Dimensões do DataFrame

```python
# dimensões do df
print("DIMENSÕES DO DATAFRAME:")
print("Linhas (Matematica):\t\t{}".format(dfmat.shape[0]))
print("Linhas (Português):\t\t{}".format(dfpor.shape[0]))
print("Colunas:\t{}".format(dfmat.shape[1]))
```

#### Transformação das colunas usadas na chave em string

```python
dfmat[["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]] = dfmat[["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]].astype(str)
dfpor[["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]] = dfpor[["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]].astype(str)
```

#### Criação de uma chave única para união das duas bases


```python
dfmat['Chave'] = dfmat["school"]+dfmat["sex"]+dfmat["age"]+dfmat["address"]+ dfmat["famsize"]+dfmat['Pstatus']+dfmat["Medu"]+dfmat['Fedu']+dfmat["Mjob"]+dfmat["Fjob"]+dfmat["reason"]+dfmat["nursery"]+dfmat["internet"]
dfpor['Chave'] = dfpor["school"]+dfpor["sex"]+dfpor["age"]+dfpor["address"]+ dfpor["famsize"]+dfpor['Pstatus']+dfpor["Medu"]+dfpor['Fedu']+dfpor["Mjob"]+dfpor["Fjob"]+dfpor["reason"]+dfpor["nursery"]+dfpor["internet"]
```

#### Criação do DataFrame Total (com full join)

```python
dftotal = pd.merge(dfmat, dfpor, on='Chave', how='outer', suffixes=('_mat','_port'), indicator=True)
dftotal.head()
```

#### Troca do Nan por vazio(''), para que apareçam nos agrupamentos

```python
dftotal[['school_mat','school_port']]=dftotal[['school_mat','school_port']].fillna('Nao faz')
```

#### Criação do DataFarame dos excluídos (só fazem uma disciplina)

```python
dfexc = dftotal[(dftotal['_merge']=='left_only') | (dftotal['_merge']=='right_only')]
dfexc.head()
```

#### Cálculo da quantidade de alunos não mesclados por disciplina

```python
index_exc_mat = dfexc[dfexc['_merge']=='left_only'].index
number_exc_mat = len(index_exc_mat)
print('O número total de alunos(as) que faz apenas matemática é: '+ str(number_exc_mat))

index_exc_port = dfexc[dfexc['_merge']=='right_only'].index
number_exc_port = len(index_exc_port)
print('O número total de alunos(as) que faz apenas portugês é: '+ str(number_exc_port))

total_exc = number_exc_mat+number_exc_port
print('O número total de alunos(as) que só faz uma das duas disciplinas é: ' + str(total_exc))
```

#### Agrupamento da base excluída pelas escolas e separação por escola


```python
dfexc_byschool = dfexc.groupby(['school_mat','school_port']).size().reset_index(name='Contagem')
dfexc_byschool
```

#### Total de alunos excluídos por escola

```python
exc_GP = dfexc_byschool[(dfexc_byschool['school_mat']=='GP') | (dfexc_byschool['school_port']=='GP')]
exc_MS = dfexc_byschool[(dfexc_byschool['school_mat']=='MS') | (dfexc_byschool['school_port']=='MS')]

total_gp = exc_GP['Contagem'].sum()
total_ms = exc_MS['Contagem'].sum()

print('O número total de alunos(as) que só faz uma das duas disciplinas no colégio GP é: '+ str(total_gp))
print('O número total de alunos(as) que só faz uma das duas disciplinas no colégio MS é: '+ str(total_ms))
```

#### Criação do DataFarame apenas dos mesclados 

```python
dfmesc = dftotal[dftotal['_merge']=='both']
```

#### Cálculo da quantidade de alunos mesclados

```python
index_mesc = dfmesc.index
number_mesc = len(index_mesc)
print('O número total de alunos(as) que faz as duas disciplinas é: '+ str(number_mesc))
```

#### Cálculo da quantidade de alunos que tiraram mais que 10 nas duas disciplinas e tem a mãe como professora

```python
df1 = dfmesc[(dfmesc['G1_mat'] > 10) & (dfmesc['G1_port'] > 10) & (dfmesc['Mjob_mat']=='teacher')]

index_df1 = df1.index
number_df1 = len(index_df1)
print('A quantidade de alunos que tiraram mais que 10 nas duas disciplinas e tem a mãe como professora é: '+str(number_df1))
df1.head()
```

#### Cálculo da quantidade de alunos que tiraram menos que 6 no G1 e G2 em uma das duas disciplinas e tem a mãe como professora

```python
df2 = dftotal[(((dftotal['G1_mat'] < 6) & (dftotal['G2_mat'] < 6)) | ((dftotal['G1_port'] < 6) & (dftotal['G2_port'] < 6))) & ((dftotal['Mjob_mat']=='teacher') | (dftotal['Mjob_port']=='teacher')) ]

index_df2 = df2.index
number_df2 = len(index_df2)

print('A quantidade de alunos que tiraram menos que 6 no G1 e G2 em uma das duas disciplinas e tem a mãe como professora é: '+str(number_df2))
df2
```

#### Alunos que tiveram ajuda extra curriular

```python
#Alunos que tem ajuda extracurricular por colégio. 
df3 = dftotal[(dftotal['schoolsup_mat']=='yes') | (dftotal['schoolsup_port']=='yes')].groupby(['school_port','school_mat']).size().reset_index(name='Total')
df3
```

```python
#Quantiade de alunos por colégio (geral)
dfschool_GP = dftotal[(dftotal['school_mat']=='GP') | (dftotal['school_port']=='GP')].groupby(['school_mat','school_port']).size().reset_index(name='Total')
dfschool_MS = dftotal[(dftotal['school_mat']=='MS') | (dftotal['school_port']=='MS')].groupby(['school_mat','school_port']).size().reset_index(name='Total')

qtd_GP = dfschool_GP['Total'].sum()
qtd_MS = dfschool_MS['Total'].sum()
```

```python
#DataFrame com alunos de cada escola que tem ajuda extracurricular 
df3_GP=df3[(df3['school_port']=='GP') | (df3['school_mat']=='GP')]
df3_MS=df3[(df3['school_port']=='MS') | (df3['school_mat']=='MS')]

#Quantidade de alunos que tiveram ajuda extra-curricular por colégio
qtd_GP3=df3_GP['Total'].sum()
qtd_MS3=df3_MS['Total'].sum()
```

```python
# Porcentagem de Alunos que tiveram ajuda extra-curricular por colégio
pct_GP3=round(qtd_GP3*100/qtd_GP,2)
pct_MS3=round(qtd_MS3*100/qtd_MS,2)

print('A porcentagem de alunos que tiveram ajuda extra curriular no colégio GP foi de: ' + str(pct_GP3)+' %')
print('A porcentagem de alunos que tiveram ajuda extra curriular no colégio MS foi de: ' + str(pct_MS3)+' %')
```
